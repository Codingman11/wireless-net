**1. Explain with your own words (1 or 2 paragraphs) the three-layer model presented in the lecture and in [[2]](https://www.ece.rutgers.edu/~marsic/books/CN/book-CN_marsic.pdf).**

**Answer:**

Three-layer model has three layers that are Link layer (Layer 1), Network layer (Layer 2), End-to-end layer (Layer 3). The bottom of the protocol stack is called Link layer which goal is to transport packets between the nodes. There is also a physical layer that is bundled with the link layer. The next upper of the protocal stack is Network layer that is responsible to bridging end hosts across intermediate links. The last one, End-to-end layer, is responsible for bringing the communication-features of the applications together.

**3. Analyze a queue system similar to the tutorial considering $\lambda = 3$ and $\mu = 4$ for 2000 time periods.** 
- Plot a histogram of the queue sizes.
- Compute the ratio of idle periods (0 elements in the queue) in relation to all time periods.

```python
#3. 

size = 2000

#Arrival
mean = 3
arrival = np.random.poisson(mean, size)

#Queue
mean = 4
service = np.random.poisson(mean, size)

queue = np.zeros(size+1)

for i in range(size):
    if ((arrival[i] - service[i]) + queue[i]) > 0:
        queue[i+1] = max(0, (arrival[i] - service[i]) + queue[i])

plt.figure(figsize=(14, 5))


plt.hist(queue)
plt.xlabel('queue size')
plt.ylabel('frequency')
plt.title('queue size distribution')
plt.grid(True)
plt.show()

#The ratio of periods

ratio = (queue == 0).sum() / size
print("Ratio of idle periods is ", (queue==0).sum(), "/", size, "=", ratio)


```
